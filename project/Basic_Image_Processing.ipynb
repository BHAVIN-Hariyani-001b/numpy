```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image,ImageFilter
```

```python
img = Image.open("./images.jpg")
img_arr = np.array(img)

plt.imshow(img_arr)
plt.title("Original image")
plt.axis('off')
plt.show()
```

```python
image_change = 1 - img_arr

plt.imshow(image_change)
plt.axis('off')
plt.title('dark image')
plt.show()
```

```python
blur_image = img.filter(ImageFilter.BLUR)
plt.imshow(blur_image)
plt.axis('off')
plt.title('blur image')
plt.show()
```

```python
height,width = 100,100
    # np.random.randint(0, 256, (height, width, 3))
    # Creates a 3D array: height  width  3 channels (RGB)
image_create = np.random.randint(0,256,(height,width,3),dtype=np.uint8)
plt.imshow(image_create)
plt.axis('off')
plt.show()
```

```python
black_image = np.random.randint(0,256,(height,width),dtype=np.uint8) # dark image
plt.imshow(black_image,cmap='grey')
plt.axis('off')
plt.show()
```
